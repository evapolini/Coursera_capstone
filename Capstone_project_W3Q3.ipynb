```python
#Get only borough with"Toronto"
mapleleaf_data =  clean_data[clean_data['Borough'].str.contains('Toronto')].reset_index(drop=True)
#mapleleaf_data['len'] = mapleleaf_data['Neighborhood'].str.len()
#mapleleaf_data.sort_values(by=['len'], inplace =True)
#mapleleaf_data = mapleleaf_data.head(1)
print (mapleleaf_data.shape )
mapleleaf_data
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#run the above function on each neighborhood and create a new dataframe called mapleleafvenues.
#mapleleafvenues= pd.DataFrame(pd.np.empty((0, 7)))   
mapleleaf_venues = getNearbyVenues(names=mapleleaf_data['Neighborhood'], latitudes=mapleleaf_data['Latitude'], longitudes=mapleleaf_data['Longitude'])
```

```python
mapleleaf_venues = mapleleaf_venues.rename(columns={'Venue Category':'VenueCategory'})
```

```python
print(mapleleaf_venues.shape)
mapleleaf_venues.head()
```

```python
mapleleaf_venues.groupby('Neighborhood').count()
print('There are {} uniques categories.'.format(len(mapleleaf_venues['VenueCategory'].unique())))
```

```python
mapleleaf_onehot = pd.get_dummies(mapleleaf_venues[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
mapleleaf_onehot['Neighborhood'] = mapleleaf_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [mapleleaf_onehot.columns[-1]] + list(mapleleaf_onehot.columns[:-1])
mapleleaf_onehot = mapleleaf_onehot[fixed_columns]

mapleleaf_onehot.head()
```

```python
mapleleaf_onehot.shape
```

```python
#Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category¶
mapleleaf_grouped = mapleleaf_onehot.groupby('Neighborhood').mean().reset_index()
mapleleaf_grouped
```

```python
#Let's print each neighborhood along with the top 5 most common venues¶
num_top_venues = 5

for hood in mapleleaf_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = mapleleaf_grouped[mapleleaf_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = mapleleaf_grouped['Neighborhood']

for ind in np.arange(mapleleaf_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(mapleleaf_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

<H1> Cluster all the neighborhood</h1>

```python
# set number of clusters
kclusters=4
mapleleaf_grouped_clustering = mapleleaf_grouped.drop('Neighborhood', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0,init='k-means++', n_init=12).fit(mapleleaf_grouped_clustering)
# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:50] 
```

```python
neighborhoods_venues_sorted.head()
```

```python
#Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
# add clustering labels
neighborhoods_venues_sorted.drop ( 'Cluster Labels',axis=1, inplace = True)
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighborhoods_venues_sorted
```

```python

mapleleaf_merged = mapleleaf_data

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
mapleleaf_merged = mapleleaf_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

mapleleaf_merged.shape
mapleleaf_merged.head()# check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=11)
kclusters

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(mapleleaf_merged['Latitude'], mapleleaf_merged['Longitude'], mapleleaf_merged['Neighborhood'], mapleleaf_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=10,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.4).add_to(map_clusters)
    
    
for lat, lon, poi, VenueCategory in zip(mapleleaf_venues['Venue Latitude'], mapleleaf_venues['Venue Longitude'], mapleleaf_venues['Neighborhood'], mapleleaf_venues['VenueCategory']):
    label = folium.Popup(str(poi) + ' VenueCategory ' + str(VenueCategory), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=1,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
map_clusters

```
